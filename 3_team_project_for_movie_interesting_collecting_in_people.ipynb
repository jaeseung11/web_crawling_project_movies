## 결론 요약

1. 크게 2가지로 나누어서 영화관련 데이터와 인구데이터로 보았을때, 영화관련 데이터(영화관 수, 좌석 수)와 선형적인 관계에 놓여 있다는 것을 알게됨. 


2. 인구데이터 측면에서 가설로는 상권이 밀집되어있고, 교통 및 인프라가 많이 밀집된 강남구 지역이 많은 양상을 띄어것이라 예상을 해었지만, 영화 데이터 수치는 맞았지만, 인구데이터 함께 비교 분석해본 결과 크게 관련을 없는 것으로 판단됨.(근거는 : HEAT MAP)


3. 특히, 2번에서 반비례 관계를 보여주고 있는 증거는 sns.pairplot, sns.scatterplot() 함수로 나타낸 그래프와 같습니다. 영화관수 와 대학 수간의 관계에서 전혀 예상과는 다른 결과를 보여짐. 그 이유는 대학교 수 및 고등학교는 교육기관이라는 명목하에 건전적인 문화라는 인식과 조경 환경을 우선시 하는 분위기가 높아서 크게 상업적이거나 이익을 추구하는 목적으로 하기에 크게 메리트가 없는것으로 판단 


4. 예전 코로나 이전에는 많은 영화관 입점이 있을것이라 판단이 되는데 년도별도 영화관의 갯수 추이 그래프 보여주었으면, 객관적인 판단하는데 도움이 됬을거라 생각이 된다. 마지막으로 영화관 목적은 대중들에게 큰 스크린으로 다함께 보는것에 취지가 있기 때문에 편중되게 있는 것이 아니라 골고루 평등하게 입지를 만들었기에 크게 눈에 띄게 편향적인 모습을 보이지 않는 것이라 추측됨.

2.영화 관객수 하나만 놓고 봤을때 매출액과 유동인구를 상관관계를 보여주는 것이 논리 비약을 보임. 그 이유는 매출액안에는 수많은 경제활동이 있기에 영화 매출액이라고 볼수 없음. 두번째 유동인구를 파악하여 영화관 입지에 사람이 많이 살고 있을거라 생각을 해보았지만. 뚜렷한 관계는 없었지만, 년도별 데이터, 동별 데이터로 가져왔으면 정확하게 파악이 됬을거라 판단. 


3.브랜드 및 입점전략도 조사도 필요.



동민이형 조언
1. 동 단위로 해야 할것.
2. 매출액에 대한 상세한 데이터에 설명 필요
3. 문제정의 하고 나서 결론을 보여준다. 이렇게 했더니 차례 데이터의 백그라운드을 보여준다.
